```python
def string_to_list(string):
    if isinstance(string, str):
        # Split the string by commas and then strip leading/trailing spaces from each element
        return [s.strip() for s in string.split(',')]
    else:
        return np.nan

# Apply the function to the 'Reserved' column
df_destinations['Reserved'] = df_destinations['Reserved'].apply(string_to_list)
df_destinations['Not Available'] = df_destinations['Not Available'].apply(string_to_list)
```

```python
# Read the text file
with open('../../data/raw/courses.txt', 'r') as file:
    # Read lines and strip newline characters
    lines = [line.strip() for line in file]

# Now 'lines' contains each element from the text file as a separate string
# Convert it into a list
courses = list(lines)
courses = sorted(courses)
print(courses)
```

```python
def determine_available(row):
    if isinstance(row['Reserved'], list):
        return row['Reserved']
    elif isinstance(row['Not Available'], list):
        print(row['Not Available'])
        return [course for course in courses if course not in row['Not Available']]
    else:
        return courses

# Apply function row-wise to create new column
df_destinations['available'] = df_destinations.apply(determine_available, axis=1)
```

```python
df_destinations.drop(['Not Available', 'Reserved'], axis = 1, inplace=True)
```

```python
def remove_string(data, columns, strings_to_remove):
    for string in strings_to_remove:

        for col in columns:
            # mask columns of the dataset that contain the string to remove
            mask = data[col].str.contains(string, na=False)

            # replace the string with an empty string
            data.loc[mask, col] = data.loc[mask, col].str.replace(string, '')

    return data
```

```python
# Open data on the best universities per subject
subjects = ['mathematics', 'computer_science']

def open_ranking_data(subjects):
    df = pd.DataFrame()

    for subject in subjects:
        dataframe_name = f'df_{subject}'

        # Read the CSV file with the correct name
        globals()[dataframe_name] = pd.read_csv(f'../../data/raw/{subject}.csv', delimiter=';', encoding_errors = 'ignore')
        print(len(globals()[dataframe_name]))

        # Create mask for rows that contain - in 2022 and 2023 column
        mask_2022 = globals()[dataframe_name]['2022'].str.contains('-', na=False)
        mask_2023 = globals()[dataframe_name]['2023'].str.contains('-', na=False)

        # Format the masked rows of 2022 and 2023 columns to include only the first ranking (ex. 651-680 --> 651)
        globals()[dataframe_name].loc[mask_2022, '2022'] = globals()[dataframe_name].loc[mask_2022, '2022'].str.split('-').str[0]
        globals()[dataframe_name].loc[mask_2023, '2023'] = globals()[dataframe_name].loc[mask_2023, '2023'].str.split('-').str[0]

        # Remove other special characters from the 2022 and 2023 columns
        globals()[dataframe_name] = remove_string(globals()[dataframe_name], ['2022', '2023'], ['=', ' '])

        # Format all numeric values such to use . as decimal separator
        columns = ['Academic', 'Employer', 'Citations', 'H', 'International', 'Score']
        globals()[dataframe_name].replace(',', '.', regex=True, inplace=True)

        # Convert columns to numeric
        columns = ['Academic', 'Employer', 'Citations', 'H', 'International', 'Score', '2022', '2023']
        globals()[dataframe_name][columns] = globals()[dataframe_name][columns].apply(pd.to_numeric)

        # Rename columns to include the subject
        columns_to_rename = ['Academic', 'Employer', 'Citations', 'H', 'International', 'Score', '2022', '2023']
        new_columns = [f'{column}_{subject}' for column in columns_to_rename]
        globals()[dataframe_name].rename(columns=dict(zip(columns_to_rename, new_columns)), inplace=True)

        # If df is empty, assign the first dataframe to it
        if df.empty:
            df = globals()[dataframe_name]
        else:
            # Merge the dataframes using the institution name as the key using NaN as fill value
            df = df.merge(globals()[dataframe_name], on='Institution', how='outer', suffixes=('', '_y'))

            # Merge location columns
            df['Location'] = df['Location'].fillna(df['Location_y'])
            df.drop('Location_y', axis=1, inplace=True)

        # assert there are no duplicates in the Institution column
        assert df['Institution'].duplicated().sum() == 0

    return df

ranking_data = open_ranking_data(subjects)
ranking_data
```

```python
# For each subject add the 2023 ranking to the df_destinations
for subject in subjects:
    column_name = f'2023_{subject}'

    # Subset the ranking data to only include the Institution and the 2023 ranking
    ranking_subset = ranking_data[['Institution', f'2023_{subject}']]

    # Merge the subset to the df_destinations
    df_destinations = df_destinations.merge(ranking_subset, left_on='University', right_on='Institution', how='left')
```

```python
# Show 10 best universities for each subject
for subject in subjects:
    column_name = f'2023_{subject}'
    print(f'Top 10 universities for {subject}')
    print(df_destinations.sort_values(by=column_name, ascending=True).head(10)[['University', 'School', column_name]])
    print('')
```
