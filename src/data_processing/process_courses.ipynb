```python
import numpy as np
import pandas as pd
```

```python
df_courses = pd.read_csv("../../data/raw/course_data.csv")
df_courses
```

```python
df_courses['Exam'] = df_courses['Exam'].str.replace('\t', '')
df_courses
```

```python
# Replace 'BIEF' with 'BIEF - fin'
df_courses['Course'] = df_courses['Course'].replace('BIEF', 'BIEF - fin')

# Duplicate rows with 'BIEF - econ'
df_dup = df_courses[df_courses['Course'] == 'BIEF - fin'].copy()
df_dup['Course'] = 'BIEF - econ'

# Concatenate original DataFrame with duplicated rows
result_df = pd.concat([df_courses, df_dup], ignore_index=True)
```

```python
result_df
```

```python
# Define a function to check if 'seminar' is in the exam column
def check_seminar(row):
    if 'seminar' in row['Exam'].lower():
        return 1
    else:
        return 0

# Create the 'seminar' column using apply
result_df['seminar'] = result_df.apply(lambda row: check_seminar(row), axis=1)
```

```python
mask = result_df['Credits'] == 1
result_df[mask]
```

```python
result_df.to_csv("../../data/processed/processed_course_data.csv")
```
